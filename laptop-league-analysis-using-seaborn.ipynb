# Data Cleaning

```python
#finding missing percentage in each cols
for col in df.columns:
    pt_missing=np.mean(df[col].isnull())
    print('{}- {}%'.format(col,pt_missing))
```

# Data Analysis

```python
# Import Imp libraries
import seaborn as sns

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure


%matplotlib inline
```

```python
df['Company'].unique()
```

```python
sns.pairplot(df,palette='husl',hue='Company')
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='Price',y='Review',data=df,hue='Company')
```

```python
df.groupby('Company')['Review'].sum().sort_values(ascending=False)
```

```python
df.groupby('Company')['Price'].max().sort_values(ascending=False)
```

```python
#count of company's laptop

plt.figure(figsize=(16, 6))
sns.countplot(data=df, x='Company',order=df['Company'].value_counts().index,palette='husl',hue='RAM')
plt.xlabel('Company')
plt.ylabel('Count')
plt.title('Count of Each Company')
plt.xticks(rotation=45)
plt.show()
```

```python
# AVG RETAIL CHARGES
df['price_diff']=df['MRP']-df['Price']
df['price_diff'].mean().round()
```

```python
#Each Company retail price difference to original price
plt.figure(figsize=(6, 8))
sns.barplot(x='price_diff',y='Company',data=df)
plt.ylabel('Company')
plt.xlabel('retail price diff')
plt.title('Retail price diff Each Company')
plt.yticks(rotation=45)
plt.show()
```
